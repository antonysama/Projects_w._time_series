```
import plotly
import plotly.express as px
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df1=pd.read_csv('man.csv',index_col='REF_DATE',parse_dates=True, dtype='unicode') # data: StatCan Table: 16-10-0048-01
df1 = df1.rename(columns={'North American Industry Classification System (NAICS)': 'naics','Seasonal adjustment': 'ssnl'})
df1 = df1.filter(['GEO','naics', 'ssnl', 'VALUE'], axis=1)
df1 = df1[df1["ssnl"].str.contains("Seasonally", na=False)]
#df1 = df1[df1["GEO"].str.contains("Alberta", na=False)] #optional/filter
df1=df1.dropna(subset=['VALUE'])
df1['VALUE']= pd.to_numeric(df1['VALUE'], errors='coerce')
#df1 = df1['2019-05-01':] #optional/filter
df1.head(1)
```

```
df2=df1.groupby(['REF_DATE','naics'])['VALUE'].agg('sum').reset_index() #optionally other groupbys can be done
df2.head(2)
```

```
fig = px.line(df2, x="REF_DATE", y="VALUE", color='naics', title="Manufacturing sales montlhy, Canada ($x000s seasonally adjusted)")
fig.show()
```
